```python
Date['date'] = pd.to_datetime(Date['date'])
```

```python
Date.info()
```

```python
Date['date_year'] = Date['date'].dt.year
```

```python
Date.sample(5)
```

```python
Date['date_month_number'] = Date['date'].dt.month
```

```python
Date.sample(5)
```

```python
Date['date_month_number'] = Date['date'].dt.month_name()
Date.sample(5)
```

```python
Date['day_of_the_month'] = Date['date'].dt.day
Date.sample(5)
```

```python
Date['day_of_week'] = Date['date'].dt.dayofweek
Date.sample(5)
```

```python
Date['day_name'] = Date['date'].dt.day_name()
Date.sample(5)
```

```python
Date['date_is_weekend'] = np.where(Date['day_name'].isin(['Sunday', 'Saturday']), 1,0)
Date.drop(columns=['product_id','city_id','orders']).head()
Date.head(5)
```

```python
Date['week_ofthe_year'] = Date['date'].dt.week
Date.sample(5)
```

```python
Date['quarter_ofthe_year'] = Date['date'].dt.quarter
Date.drop(columns=['product_id','city_id','orders']).head()
Date.sample(5)
```

```python
Date['semester'] = np.where(Date['quarter_ofthe_year'].isin([1,2]), 1, 2)
Date.drop(columns=['product_id','city_id','orders']).head()
Date.sample(5)
```

```python
import datetime
today = datetime.datetime.today
today()
```

```python
today() - Date['date']
```

```python
(today() - Date['date']).dt.days
```

```python
np.round((today() - Date['date']) / np.timedelta64(1, 'M'),0)
```
