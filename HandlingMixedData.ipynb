```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head(5)
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot.bar()
fig.set_title('passenger travelling with')
```

```python
df['numer_numerical'] = pd.to_numeric(df['number'], errors= 'coerce', downcast='integer')
```

```python
df['numer_categorical'] = np.where(df['number'].isnull(),df['number'], np.nan)
```

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
df['Cabin_num'] = df['Cabin'].str.extract('(\d+)') #this is to extract the numerical part
df['cabin_cat'] = df['Cabin'].str[0] #this captures the first letter e.g. C, B 
```

```python
df.head(5)
```

```python
df['cabin_cat'].value_counts()
```

```python
df['cabin_cat'].value_counts().plot(kind='bar')
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda s: s.split()[-1])
df['ticket_num'] = pd.to_numeric(df['ticket_num'],
                                   errors='coerce',
                                   downcast='integer')

# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s: s.split()[0])
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan,
                              df['ticket_cat'])

df.head(20)
```
